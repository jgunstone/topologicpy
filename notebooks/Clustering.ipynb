```python
# This cell is not needed if you have pip installed topologicpy
import sys
sys.path.append("C:/Users/sarwj/OneDrive - Cardiff University/Documents/GitHub/topologicpy/src")
from topologicpy.Vertex import Vertex
from topologicpy.Edge import Edge
from topologicpy.Wire import Wire
from topologicpy.Face import Face
from topologicpy.Shell import Shell
from topologicpy.Cell import Cell
from topologicpy.CellComplex import CellComplex
from topologicpy.Cluster import Cluster
from topologicpy.Topology import Topology
from topologicpy.IFC import IFC
from topologicpy.Dictionary import Dictionary
from topologicpy.Helper import Helper
```

```python
r1 = Face.Rectangle(width=2)
r2 = Face.Rectangle(width=5, length=1)
r2 = Topology.Translate(r2, 5, 0, 0)
r3 = Face.Rectangle(width=3, length=3)
r3 = Topology.Translate(r3, 1.5, -2.2, 0)
# Let's add connectors
r4 = Face.Rectangle(origin=Vertex.ByCoordinates(0.75,-0.5,0), width=0.2, length=0.75)
r5 = Face.Rectangle(origin=Vertex.ByCoordinates(2.75,-0.5,0), width=0.2, length=0.75)
everything = Topology.Show(Cluster.ByTopologies([r1, r2, r3, r4, r5])
Topology.Show(r1, r2, r3, r4, r5)
```

```python
rectangles = [r1, r2, r3]
points = []
for r in rectangles:
    t = Topology.Triangulate(r, mode=1)
    verts = Topology.Vertices(t)
    points.extend(verts)

Topology.Show(points, vertexSize=8, backgroundColor="white")
```

```python
#clusters = Cluster.KMeans(points, k=3, keys=['x','y'])
clusters = Cluster.DBSCAN(points, keys=['x', 'y'], epsilon = 0.6, minSamples = 3)
clusters = clusters[0]
n = len(clusters)
print(n)
print(clusters)
colors = ["red", "green", "blue", "yellow", "pink", "orange", "purple"]
centroids = []
for i, cluster in enumerate(clusters):
    # Optional
    c = Topology.Centroid(cluster)
    d = Dictionary.ByKeyValue("color", "black")
    c = Topology.SetDictionary(c, d)
    centroids.append(c)

    verts = Topology.Vertices(cluster)
    d = Dictionary.ByKeyValue("color", colors[i])
    for v in verts:
        v = Topology.SetDictionary(v, d)
```

```python
Topology.Show(clusters, centroids, rectangles, vertexColorKey="color", vertexSize=16, backgroundColor="white")
```
